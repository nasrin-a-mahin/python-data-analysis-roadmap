# 1Ô∏è‚É£ Import Required Libraries

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

# 2Ô∏è‚É£ Load Dataset

```python


df = pd.read_csv("datasets/student_Marks.csv")

df.head()

```


## 3Ô∏è‚É£ Basic Exploration

```python


print("Shape:", df.shape)
print("\nColumns:")
print(df.columns)

df.info()

df.describe()
```

## 4Ô∏è‚É£ Check Missing Values & Duplicates

```python



df.isnull().sum()

df.duplicated().sum()
```

## 5Ô∏è‚É£ Distribution of Marks

```python



plt.figure(figsize=(8,5))
sns.histplot(df['Marks'], bins=15, kde=True)

plt.title("Distribution of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

## 6Ô∏è‚É£ Study Time vs Marks

```python



plt.figure(figsize=(8,5))
sns.scatterplot(
    x='time_study',
    y='Marks',
    data=df
)

plt.title("Study Time vs Marks")
plt.xlabel("Study Time (hours)")
plt.ylabel("Marks")

plt.show()
```

## 7Ô∏è‚É£ Regression Plot (Study Time ‚Üí Marks)

```python



plt.figure(figsize=(8,5))
sns.regplot(
    x='time_study',
    y='Marks',
    data=df
)

plt.title("Regression: Study Time vs Marks")
plt.show()
```

## 8Ô∏è‚É£ Number of Courses vs Marks

```python



plt.figure(figsize=(8,5))
sns.boxplot(
    x='number_courses',
    y='Marks',
    data=df
)

plt.title("Marks by Number of Courses")
plt.xlabel("Number of Courses")
plt.ylabel("Marks")

plt.show()
```

## 9Ô∏è‚É£ Average Marks by Number of Courses

```python



course_avg = df.groupby('number_courses')['Marks'].mean()
course_avg
```

## üîü Correlation Analysis

```python



correlation = df.corr()
correlation

plt.figure(figsize=(6,4))
sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()
```

## 1Ô∏è‚É£1Ô∏è‚É£ Statistical Measures


```python


mean_marks = df['Marks'].mean()
median_marks = df['Marks'].median()
std_marks = df['Marks'].std()

mean_marks, median_marks, std_marks
```

## 1Ô∏è‚É£2Ô∏è‚É£ High & Low Performers

```python



# Top 5 students
top_students = df.sort_values(by='Marks', ascending=False).head(5)
top_students

# Bottom 5 students
low_students = df.sort_values(by='Marks').head(5)
low_students
```

## 1Ô∏è‚É£3Ô∏è‚É£ Insights

‚úî Study time has a **strong positive correlation** with marks  
‚úî Number of courses alone does not guarantee higher marks  
‚úî Students studying more consistently score higher  
‚úî Marks distribution is right-skewed  
‚úî Some students perform well even with fewer courses



## 1Ô∏è‚É£4Ô∏è‚É£ Conclusion

This analysis shows that **time invested in studying matters more than course load**.
Effective study habits lead to higher academic performance.
