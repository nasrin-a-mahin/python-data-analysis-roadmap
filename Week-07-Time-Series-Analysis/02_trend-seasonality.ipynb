```python
import pandas as pd
import matplotlib.pyplot as plt

dates = pd.date_range(start='2024-01-01', periods=12, freq='M')
sales = [100,120,130,150,170,180,200,220,210,230,250,270]

ts = pd.Series(sales, index=dates)
ts
```

```python
# Plot time series
plt.figure()
plt.plot(ts)
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```
