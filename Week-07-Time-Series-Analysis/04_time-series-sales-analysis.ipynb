### Business Objective
- Analyze sales over time
- Identify trends and patterns


```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'date': pd.date_range('2023-01-01', periods=12, freq='M'),
    'sales': [200,220,210,230,250,270,260,280,300,320,310,330]
}

df = pd.DataFrame(data)
df.set_index('date', inplace=True)
df
```

```python
# Plot sales trend
plt.figure()
plt.plot(df.index, df['sales'])
plt.title('Monthly Sales Analysis')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

### Key Insights
- Sales show an upward trend
- Best performance in later months
- Useful for forecasting and planning
