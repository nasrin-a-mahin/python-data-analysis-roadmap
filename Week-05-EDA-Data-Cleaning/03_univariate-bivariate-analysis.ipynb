## Objective

Analyze individual variables and relationships between variables to identify patterns, drivers, and anomalies.



1Ô∏è‚É£ Import Libraries & Load Clean Data


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("retail_sales_cleaned.csv")
```


üîπ PART 1: UNIVARIATE ANALYSIS
2Ô∏è‚É£ Numerical Variables
Sales Distribution


```python

sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()

df['Sales'].describe()
```



üìå Insight (Markdown)

Sales is right-skewed, indicating most orders are low-to-medium value.

Profit Distribution


```python
sns.boxplot(x=df['Profit'])
plt.title("Profit Distribution")
plt.show()

df['Profit'].describe()

```


üìå Insight

Negative profits exist, possibly due to high discounts or returns.

3Ô∏è‚É£ Categorical Variables
Product Category Frequency


```python
df['Product_Category'].value_counts().plot(kind='bar')
plt.title("Orders by Product Category")
plt.show()
```


Region Distribution


```python
df['Region'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Customer Distribution by Region")
plt.show()

```


üìå Analyst Note

Pie charts are useful for distribution, not comparison.

üîπ PART 2: BIVARIATE ANALYSIS
4Ô∏è‚É£ Numerical vs Numerical
Sales vs Profit


```python
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.show()

df[['Sales', 'Profit']].corr()

```


üìå Insight

Positive correlation exists, but variability increases at higher sales.

5Ô∏è‚É£ Categorical vs Numerical
Category vs Sales


```python
sns.boxplot(x='Product_Category', y='Sales', data=df)
plt.xticks(rotation=45)
plt.show()
```


Region vs Profit


```python
sns.barplot(x='Region', y='Profit', data=df, estimator='mean')
plt.xticks(rotation=45)
plt.show()

```


üìå Insight

Certain regions show consistently higher average profits.

6Ô∏è‚É£ Categorical vs Categorical
Gender vs Product Category


```python
pd.crosstab(df['Gender'], df['Product_Category'])

sns.heatmap(pd.crosstab(df['Gender'], df['Product_Category']),
            annot=True, fmt='d', cmap='Blues')
plt.title("Gender vs Product Category")
plt.show()
```


7Ô∏è‚É£ Discount Impact Analysis


```python
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```



üìå Critical Insight

Higher discounts often lead to reduced or negative profits.

8Ô∏è‚É£ Summary of Key Findings (Markdown)
1. Sales and profit are positively correlated
2. Technology category generates higher sales
3. High discounts negatively impact profitability
4. Certain regions outperform others
