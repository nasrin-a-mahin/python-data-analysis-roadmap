3Ô∏è‚É£ Load Clean Dataset


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("retail_sales_cleaned.csv")
```


4Ô∏è‚É£ KPI Overview


```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
avg_discount = df['Discount'].mean()

total_sales, total_profit, avg_discount

```


üìå Executive Summary

- Total Revenue: $XXM
- Total Profit: $XM
- Average Discount: XX%



5Ô∏è‚É£ Insight 1: Product Category Performance


```python
category_perf = df.groupby('Product_Category')[['Sales','Profit']].sum()
category_perf.sort_values('Sales', ascending=False)

category_perf.plot(kind='bar', figsize=(10,5))
plt.title("Sales & Profit by Category")
plt.show()

```


üìå Insight

Technology category leads both sales and profit.



6Ô∏è‚É£ Insight 2: Discount vs Profit


```python
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Impact of Discount on Profit")
plt.show()

```


üìå Insight

Orders with discounts above 30% frequently result in losses.



7Ô∏è‚É£ Insight 3: Regional Performance


```python
region_perf = df.groupby('Region')['Profit'].mean().sort_values(ascending=False)
region_perf

sns.barplot(x=region_perf.index, y=region_perf.values)
plt.xticks(rotation=45)
plt.title("Average Profit by Region")
plt.show()

```


üìå Insight

West region shows the highest average profitability.



8Ô∏è‚É£ Insight 4: Customer Age & Spending


```python
sns.boxplot(x='Age', y='Sales', data=df)
plt.title("Age vs Sales")
plt.show()
```



üìå Insight

Customers aged 30‚Äì45 contribute higher median sales.
