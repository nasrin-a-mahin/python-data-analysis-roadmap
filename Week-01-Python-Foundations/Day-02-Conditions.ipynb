ðŸ§  What are Conditional Statements?

Conditional statements allow a program to make decisions based on conditions.

ðŸ”¹ 1. The if Statement
Syntax:

if condition:
    # code block


```python
age = 18

if age >= 18:
    print("You are eligible to vote")
```

ðŸ”¹ 2. The ifâ€“else Statement

```python
age = 15

if age >= 18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

3. The ifâ€“elifâ€“else Statement

Used when multiple conditions are needed.

```python
marks = 82

if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 60:
    grade = "C"
else:
    grade = "Fail"

print("Marks:", marks)
print("Grade:", grade)
```

ðŸ”¹ 4. Comparison Operators

| Operator | Meaning               |
| -------- | --------------------- |
| `==`     | Equal to              |
| `!=`     | Not equal             |
| `>`      | Greater than          |
| `<`      | Less than             |
| `>=`     | Greater than or equal |
| `<=`     | Less than or equal    |


```python
a = 10
b = 20

print(a == b)
print(a != b)
print(a < b)
print(a >= b)
```

ðŸ”¹ 5. Logical Operators

| Operator | Description                  |
| -------- | ---------------------------- |
| `and`    | Both conditions must be true |
| `or`     | At least one condition true  |
| `not`    | Reverses the condition       |


```python
age = 22
has_id = True

if age >= 18 and has_id:
    print("Entry allowed")
else:
    print("Entry denied")
```

ðŸ”¹ 6. Nested Conditions

```python
marks = 85

if marks >= 60:
    if marks >= 90:
        print("Excellent")
    else:
        print("Good")
else:
    print("Needs Improvement")
```

ðŸ”¹ 7. Using input() with Conditions

```python
temperature = float(input("Enter temperature: "))

if temperature > 30:
    print("It's a hot day")
elif temperature >= 20:
    print("Weather is pleasant")
else:
    print("It's cold")
```

ðŸ”¹ 8. Real-World Example: Even or Odd

```python
number = int(input("Enter a number: "))

if number % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```
