üîπ 1. Lists
What is a List?

A list is an ordered, changeable (mutable) collection of items.

Creating a List

```python
marks = [78, 85, 90, 66, 88]
print(marks)
```

üîπ 2. Accessing List Elements (Indexing)

‚ö†Ô∏è Indexing starts from 0

```python
marks = [78, 85, 90, 66, 88]

print(marks[0])    # First element
print(marks[2])    # Third element
print(marks[-1])   # Last element
```

üîπ 3. Modifying List Elements

```python
marks = [78, 85, 90]

marks[1] = 88
print(marks)
```

üîπ 4. Common List Methods

```python
numbers = [10, 20, 30]

numbers.append(40)      # Add element
numbers.insert(1, 15)   # Insert at index
numbers.remove(20)      # Remove value
numbers.pop()           # Remove last element

print(numbers)
```

üîπ 5. List Length, Sum, Min, Max

```python
marks = [78, 85, 90, 66, 88]

print("Total Subjects:", len(marks))
print("Total Marks:", sum(marks))
print("Highest Marks:", max(marks))
print("Lowest Marks:", min(marks))
print("Average Marks:", sum(marks) / len(marks))
```

üîπ 6. Looping Through a List

```python
subjects = ["Math", "Science", "English"]

for subject in subjects:
    print(subject)
```

üîπ 7. Tuples
What is a Tuple?

A tuple is an ordered, unchangeable (immutable) collection of items.

```python
subjects = ("Math", "Science", "English")
print(subjects)
```

üîπ 8. Accessing Tuple Elements

‚ö†Ô∏è Tuples cannot be modified:

```python
subjects = ("Math", "Science", "English")

print(subjects[0])
print(subjects[-1])
```

üîπ 9. List vs Tuple

| Feature     | List         | Tuple      |
| ----------- | ------------ | ---------- |
| Mutable     | Yes          | No         |
| Syntax      | `[ ]`        | `( )`      |
| Performance | Slower       | Faster     |
| Use Case    | Data changes | Fixed data |


üîπ 10. Real-World Example: Student Marks Analysis

```python
marks = [78, 85, 90, 66, 88]

for mark in marks:
    if mark >= 75:
        print(mark, "Good")
    else:
        print(mark, "Needs Improvement")
```

üîπ 11. Mini Practice Programs
‚úÖ Program 1: Find Even Numbers in a List

```python
numbers = [10, 15, 20, 25, 30]

for num in numbers:
    if num % 2 == 0:
        print(num)
```

‚úÖ Program 2: Convert List to Tuple

```python
cities = ["Delhi", "Mumbai", "Chennai"]
cities_tuple = tuple(cities)

print(cities_tuple)
```
