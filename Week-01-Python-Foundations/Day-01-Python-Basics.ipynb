üîπ 3. Data Types

Python automatically detects data types.

```python
name = "Alice"      # String
age = 20            # Integer
score = 85.75       # Float
is_student = True   # Boolean

print(type(name))
print(type(age))
print(type(score))
print(type(is_student))
```

Common Data Types

| Type  | Example        |
| ----- | -------------- |
| int   | `10`           |
| float | `3.14`         |
| str   | `"Python"`     |
| bool  | `True / False` |


üîπ 4. Basic Arithmetic Operations

```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

üîπ 5. String Operations

```python
first_name = "John"
last_name = "Doe"

# Concatenation
full_name = first_name + " " + last_name
print(full_name)

# Repetition
print(first_name * 3)
```

üîπ 8. Simple Mini Practice
üìå Program: Calculate Total and Average Marks

```python
math = int(input("Enter Math marks: "))
science = int(input("Enter Science marks: "))
english = int(input("Enter English marks: "))

total = math + science + english
average = total / 3

print("Total Marks:", total)
print("Average Marks:", average)
```
