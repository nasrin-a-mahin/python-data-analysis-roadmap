ðŸ§  What is a Function?

A function is a block of reusable code that performs a specific task.
Instead of writing the same code multiple times, you can call a function.

```python
#Syntax:

def function_name(parameters):
    # code block
    return value  # optional
```

ðŸ”¹ 1. Defining and Calling a Function

```python
def greet():
    print("Hello, welcome to Python!")

# Call the function
greet()
```

ðŸ”¹ 2. Function with Parameters

Parameters allow you to pass data to a function.

```python
def greet_user(name):
    print("Hello", name, "Welcome!")

greet_user("Amit")
greet_user("Neha")
```

ðŸ”¹ 3. Function with Return Value

```python
def add(a, b):
    return a + b

result = add(10, 20)
print("Sum:", result)
```

ðŸ”¹ 4. Multiple Parameters

```python
def student_info(name, age, marks):
    print(f"Name: {name}, Age: {age}, Marks: {marks}")

student_info("Amit", 20, 88)
```

ðŸ”¹ 5. Default Parameters

```python
def greet(name="Student"):
    print("Hello", name)

greet()          # Uses default
greet("Neha")    # Overrides default
```

ðŸ”¹ 6. Keyword Arguments

```python
def student_info(name, age, marks):
    print(f"Name: {name}, Age: {age}, Marks: {marks}")

student_info(age=21, marks=90, name="Rahul")
```

ðŸ”¹ 7. Variable-Length Arguments (*args & **kwargs)

```python
# *args for multiple positional arguments
def total_marks(*marks):
    return sum(marks)

print(total_marks(80, 90, 85))

# **kwargs for multiple keyword arguments
def student_details(**details):
    for key, value in details.items():
        print(key, ":", value)

student_details(name="Neha", age=20, marks=92)
```

ðŸ”¹ 8. Scope of Variables

```python
x = 10  # global variable

def test():
    y = 5  # local variable
    print("Inside function:", x + y)

test()
print("Outside function:", x)
# print(y)  âŒ Error: y is local
```

ðŸ”¹ 9. Real-World Example: Grade Calculator

```python
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "Fail"

marks_list = [88, 92, 67, 55]

for marks in marks_list:
    print(marks, "->", calculate_grade(marks))
```

ðŸ”¹ 10. Mini Practice Programs

```python
# âœ… Program 1: Factorial Using Function

def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

print(factorial(5))  # 120

# âœ… Program 2: Check Even or Odd

def check_even(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(check_even(10))
print(check_even(15))

# âœ… Program 3: Sum of List Elements

def sum_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

marks = [78, 85, 90]
print("Total Marks:", sum_list(marks))
```
