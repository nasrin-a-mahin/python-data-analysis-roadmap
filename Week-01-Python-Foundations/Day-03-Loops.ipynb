üß† What are Loops?

Loops allow you to repeat a block of code multiple times without writing it again and again.

```python
#Example without loop ‚ùå:

print("Hello")
print("Hello")
print("Hello")

```

```python
#With loop ‚úÖ:

for i in range(3):
    print("Hello")
```

üîπ 1. The for Loop
for variable in sequence:
    # code block

```python
for i in range(5):
    print(i)

#üëâ range(5) generates numbers: 0, 1, 2, 3, 4
```

üîπ 2. range() Function

range(start, stop, step)


```python
for i in range(1, 6):
    print(i)

for i in range(1, 10, 2):
    print(i)

```

üîπ 3. Looping Through a List

```python
subjects = ["Math", "Science", "English"]

for subject in subjects:
    print(subject)
```

üîπ 4. The while Loop

while condition:
    # code block


```python
count = 1

while count <= 5:
    print(count)
    count += 1



#‚ö†Ô∏è Always update the condition, otherwise you get an infinite loop.
```

üîπ 5. break Statement

Stops the loop immediately.

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

üîπ 6. continue Statement

Skips the current iteration.

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

üîπ 7. Nested Loops

```python
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```

üîπ 8. Real-World Example: Sum of Numbers

```python
total = 0

for i in range(1, 6):
    total += i

print("Sum:", total)
```

üîπ 9. Using while with User Input

```python
password = ""

while password != "python":
    password = input("Enter password: ")

print("Access granted")
```

üîπ 10. Mini Practice Programs
‚úÖ Program 1: Print Even Number

```python
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

‚úÖ Program 2: Multiplication Table

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

üìù Key Takeaways
