ðŸ† Advanced Mini-Project: Student Result Analyzer with CSV
ðŸŽ¯ Project Objectives

Store student details (name, variable subjects, marks)

Calculate total, average, grades

Display results

Save results to a CSV file

Read and analyze student data from CSV

Find toppers and lowest scorers

Make project reusable with functions

```python
# ðŸ”¹ Step 1: Import Libraries

import csv

# ðŸ”¹ Step 2: Define Functions

# Function 1: Calculate total marks

def calculate_total(marks):
    return sum(marks)

# Function 2: Calculate average

def calculate_average(marks):
    return sum(marks) / len(marks)

# Function 3: Assign grade

def assign_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 60:
        return "C"
    else:
        return "Fail"

# Function 4: Display student result

def display_result(student):
    print("\nStudent Name:", student["name"])
    print("Subjects & Marks:", student["marks"])
    print("Total Marks:", student["total"])
    print("Average:", student["average"])
    print("Grade:", student["grade"])
    print("-" * 40)

# Function 5: Save results to CSV

def save_to_csv(filename, students, subjects):
    with open(filename, mode="w", newline="") as file:
        writer = csv.writer(file)
        header = ["Name"] + subjects + ["Total", "Average", "Grade"]
        writer.writerow(header)
        
        for student in students:
            row = [student["name"]] + student["marks"] + [student["total"], student["average"], student["grade"]]
            writer.writerow(row)

# ðŸ”¹ Step 3: Input Subjects and Student Data

students = []

# Input subjects
subjects = []
num_subjects = int(input("Enter number of subjects: "))
for i in range(num_subjects):
    subject_name = input(f"Enter name of subject {i+1}: ")
    subjects.append(subject_name)

# Input students
num_students = int(input("\nEnter number of students: "))
for i in range(num_students):
    print(f"\nEnter details for student {i+1}:")
    name = input("Name: ")
    
    marks = []
    for subject in subjects:
        mark = int(input(f"Marks in {subject}: "))
        marks.append(mark)
    
    total = calculate_total(marks)
    avg = calculate_average(marks)
    grade = assign_grade(avg)
    
    student_data = {
        "name": name,
        "marks": marks,
        "total": total,
        "average": avg,
        "grade": grade
    }
    
    students.append(student_data)
    display_result(student_data)

# ðŸ”¹ Step 4: Display All Students Summary

print("\n=== All Students Summary ===")
for student in students:
    print(f"{student['name']} -> Total: {student['total']}, Average: {student['average']:.2f}, Grade: {student['grade']}")

# ðŸ”¹ Step 5: Find Topper and Lowest Scorer

topper = max(students, key=lambda x: x["average"])
lowest = min(students, key=lambda x: x["average"])

print("\nTopper:", topper["name"], "with average", topper["average"])
print("Lowest scorer:", lowest["name"], "with average", lowest["average"])

# ðŸ”¹ Step 6: Search Student by Name

search_name = input("\nEnter student name to search: ").strip()
found = False
for student in students:
    if student["name"].lower() == search_name.lower():
        display_result(student)
        found = True
        break
if not found:
    print("Student not found!")

# ðŸ”¹ Step 7: Save Results to CSV

filename = "student_results.csv"
save_to_csv(filename, students, subjects)
print(f"\nAll student results saved to {filename}")

# ðŸ”¹ Step 8: Extra Analysis (Optional)


# Average per subject

subject_totals = [0] * len(subjects)
for student in students:
    for i, mark in enumerate(student["marks"]):
        subject_totals[i] += mark

print("\nAverage per subject:")
for i, total in enumerate(subject_totals):
    print(f"{subjects[i]}: {total / len(students):.2f}")


# List all students who failed

failed_students = [s["name"] for s in students if s["grade"] == "Fail"]
print("\nStudents who failed:", failed_students)
```
