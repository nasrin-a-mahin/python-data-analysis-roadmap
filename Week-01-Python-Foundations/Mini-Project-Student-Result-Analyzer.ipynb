ðŸŽ¯ Project Objectives

=>Store student details (name, marks)

=>Calculate total, average marks

=>Assign grades

=>Display results in a readable format

ðŸ”¹ Step 1: Plan the Project

Data we need:

Student name

Marks for 3 subjects (Math, Science, English)

Tasks:

Input student data

Calculate total marks and average

Assign grade based on average

Display student result

Repeat for multiple students

ðŸ”¹ Step 2: Define Functions

```python
# Function 1: calculate_total

def calculate_total(marks):
    return sum(marks)

# Function 2: calculate_average

def calculate_average(marks):
    return sum(marks) / len(marks)

# Function 3: assign_grade

def assign_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 60:
        return "C"
    else:
        return "Fail"

# Function 4: display_result

def display_result(name, marks, total, avg, grade):
    print("\nStudent Name:", name)
    print("Marks:", marks)
    print("Total Marks:", total)
    print("Average:", avg)
    print("Grade:", grade)
    print("-" * 30)
```

ðŸ”¹ Step 3: Input Student Data Using Loops

```python
students = []  # List to store all student dictionaries

num_students = int(input("Enter number of students: "))

for i in range(num_students):
    print(f"\nEnter details for student {i+1}:")
    name = input("Name: ")
    
    marks = []
    for subject in ["Math", "Science", "English"]:
        mark = int(input(f"Marks in {subject}: "))
        marks.append(mark)
    
    total = calculate_total(marks)
    avg = calculate_average(marks)
    grade = assign_grade(avg)
    
    student_data = {
        "name": name,
        "marks": marks,
        "total": total,
        "average": avg,
        "grade": grade
    }
    
    students.append(student_data)
    
    display_result(name, marks, total, avg, grade)






```

ðŸ”¹ Step 4: Display All Students Summary

```python
# ðŸ”¹ Step 4: Display All Students Summary


print("\n=== All Students Summary ===")
for student in students:
    print(f"{student['name']} -> Total: {student['total']}, Average: {student['average']:.2f}, Grade: {student['grade']}")
```

ðŸ”¹ Step 5: Extra Features (Optional)

```python
# ðŸ”¹ Step 5: Extra Features (Optional)
# Highest and Lowest Average


highest = max(students, key=lambda x: x["average"])
lowest = min(students, key=lambda x: x["average"])

print("\nTopper:", highest["name"], "with average", highest["average"])
print("Lowest scorer:", lowest["name"], "with average", lowest["average"])

 
# Search by Student Name


search_name = input("\nEnter name to search: ")
found = False
for student in students:
    if student["name"].lower() == search_name.lower():
        display_result(student["name"], student["marks"], student["total"], student["average"], student["grade"])
        found = True
        break
if not found:
    print("Student not found!")
```
