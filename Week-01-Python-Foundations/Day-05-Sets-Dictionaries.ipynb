ðŸ§  Why Sets & Dictionaries?

Sets store unique items (no duplicates).
Useful when you need to remove duplicates or compare collections.

Dictionaries store key-value pairs.
Useful for structured data, e.g., student marks, employee info.

ðŸ”¹ 1. Sets
What is a Set?

A set is an unordered collection of unique elements.

```python
fruits = {"apple", "banana", "cherry", "apple"}
print(fruits)  # Duplicates removed
```

ðŸ”¹ 2. Creating Sets

```python
# From a list
numbers = set([1, 2, 2, 3, 4])
print(numbers)

# Empty set
empty_set = set()
print(empty_set)
```

ðŸ”¹ 3. Basic Set Operations

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union
print("Union:", A | B)  # or A.union(B)

# Intersection
print("Intersection:", A & B)  # or A.intersection(B)

# Difference
print("Difference (A-B):", A - B)
```

ðŸ”¹ 4. Adding & Removing Elements

```python
fruits = {"apple", "banana"}

# Add
fruits.add("orange")
print(fruits)

# Remove
fruits.remove("banana")  # Error if item not present
print(fruits)

# Discard (no error if missing)
fruits.discard("banana")
print(fruits)
```

ðŸ”¹ 5. Looping Through a Set

```python
fruits = {"apple", "banana", "cherry"}

for fruit in fruits:
    print(fruit)
```

âš ï¸ Sets are unordered â€“ elements may appear in any order.

ðŸ”¹ 6. Dictionaries
What is a Dictionary?

A dictionary is a collection of key-value pairs.

```python
student = {
    "name": "Alice",
    "age": 20,
    "marks": 88
}

print(student)
```

ðŸ”¹ 7. Accessing Dictionary Items

```python
print(student["name"])
print(student.get("marks"))
```

ðŸ”¹ 8. Modifying Dictionary Items

```python
student["marks"] = 90  # Update value
student["grade"] = "A"  # Add new key-value

print(student)
```

ðŸ”¹ 9. Removing Dictionary Items

```python
# Remove specific key
student.pop("grade")
print(student)

# Remove last inserted item
student.popitem()
print(student)

# Clear dictionary
# student.clear()
# print(student)
```

ðŸ”¹ 10. Looping Through a Dictionary

```python
student = {"name": "Alice", "age": 20, "marks": 88}

# Loop through keys
for key in student:
    print(key, "->", student[key])

# Loop through items
for key, value in student.items():
    print(key, ":", value)
```

ðŸ”¹ 11. Real-World Example: Student Grades

```python
students = {
    "Amit": 85,
    "Neha": 92,
    "Rahul": 74,
    "Sneha": 67,
    "Priya": 58
}

# Print each student with grade
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "Fail"

for name, marks in students.items():
    print(f"{name}: {marks} -> {calculate_grade(marks)}")
```
