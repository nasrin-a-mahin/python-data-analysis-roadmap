## ðŸ“ Day-10: NumPy Operations

```python
import numpy as np
```

1ï¸âƒ£ Element-wise Arithmetic Operations
Problem

Given arrays a = np.array([1, 2, 3]) and b = np.array([4, 5, 6]), perform addition, subtraction, multiplication, division, and power.


```python

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Power:", a ** 2)

```

âœ… Element-wise operations do not require loops.

2ï¸âƒ£ Universal Functions (ufuncs)

NumPy has built-in fast mathematical functions.

```python

arr = np.array([1, 4, 9, 16])

print("Square root:", np.sqrt(arr))
print("Exponential:", np.exp(arr))
print("Natural log:", np.log(arr))
print("Sine:", np.sin(arr))
```

3ï¸âƒ£ Aggregation Functions

Compute statistics over the whole array or along an axis.


```python

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Sum:", np.sum(arr))          # Sum all elements
print("Sum along axis 0:", np.sum(arr, axis=0))  # Column-wise
print("Sum along axis 1:", np.sum(arr, axis=1))  # Row-wise

print("Mean:", np.mean(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))
print("Standard deviation:", np.std(arr))
```

4ï¸âƒ£ Axis-wise Operations

axis=0 â†’ operation along columns

axis=1 â†’ operation along rows

```python



arr = np.array([[1,2,3],[4,5,6]])

print("Column sum (axis=0):", np.sum(arr, axis=0))  # [5 7 9]
print("Row sum (axis=1):", np.sum(arr, axis=1))     # [6 15]
```

5ï¸âƒ£ Boolean Operations

Create mask arrays for filtering

```python



arr = np.array([10, 15, 20, 25])
mask = arr > 15
print("Mask:", mask)        # [False False True True]
print("Filtered:", arr[mask])  # [20 25]
```

Count values satisfying condition

```python


print("Count > 15:", np.sum(arr > 15))
```

6ï¸âƒ£ Broadcasting

NumPy allows operations between arrays of different shapes, following broadcasting rules.


```python


arr = np.array([[1,2,3],[4,5,6]])
scalar = 10

print("Add scalar:", arr + scalar)  # Adds 10 to every element

row = np.array([1,0,1])
print("Add row:", arr + row)        # Row broadcast across each row
```



âœ… Very powerful for real-world numerical computations.

7ï¸âƒ£ Linear Algebra Operations

Dot product, transpose, determinant, inverse (if square)

```python


A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

# Matrix multiplication
print("Dot product:\n", np.dot(A,B))

# Transpose
print("Transpose:\n", A.T)

# Inverse
print("Inverse:\n", np.linalg.inv(A))

# Determinant
print("Determinant:", np.linalg.det(A))
```

8ï¸âƒ£ Cumulative Operations

```python


arr = np.array([1,2,3,4])

print("Cumulative sum:", np.cumsum(arr))   # [1 3 6 10]
print("Cumulative product:", np.cumprod(arr)) # [1 2 6 24]

```

9ï¸âƒ£ Sorting & Unique

```python

arr = np.array([3,1,4,1,5,9,2])
print("Sorted:", np.sort(arr))
print("Unique elements:", np.unique(arr))
```


10ï¸âƒ£ Advanced Operations (Optional)

Clip values: restrict range

```python


arr = np.array([1, 5, 10, 15])
print("Clipped [3,12]:", np.clip(arr, 3, 12))  # [3 5 10 12]


```

Where function: conditional replacement

```python


arr = np.array([1,2,3,4])
print(np.where(arr%2==0, "Even", "Odd"))
# ['Odd' 'Even' 'Odd' 'Even']
```
