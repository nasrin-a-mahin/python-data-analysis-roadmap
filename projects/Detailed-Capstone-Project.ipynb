3Ô∏è‚É£ Tools & Libraries Used


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```


4Ô∏è‚É£ Step-by-Step Analysis (REAL ANALYST WORK)
üîπ Step 1: Data Loading & Inspection


```python
df = pd.read_csv("datasets/sales_data.csv")
df.head()

df.info()
df.describe()

```


üìå Insights:

Dataset contains sales transactions

Revenue column present

Date needs conversion

üîπ Step 2: Data Cleaning

‚úî Convert date
‚úî Handle missing values
‚úî Remove duplicates


```python

df['Order_Date'] = pd.to_datetime(df['Order_Date'])

df.isnull().sum()

df.drop_duplicates(inplace=True)
```



üìå Business Reasoning:
Clean data ensures accurate revenue reporting.

üîπ Step 3: Feature Engineering


```python
df['Month'] = df['Order_Date'].dt.month
df['Year'] = df['Order_Date'].dt.year

```


üìå Enables trend & seasonal analysis.

üîπ Step 4: Exploratory Data Analysis (EDA)
üî∏ Total Revenue


```python
total_revenue = df['Revenue'].sum()
total_revenue
```



üìå Insight:
Shows overall business performance.

üî∏ Revenue by Region


```python
region_revenue = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
region_revenue

region_revenue.plot(kind='bar', title='Revenue by Region')
plt.show()

```


üìå Business Insight:
Top regions drive majority of revenue ‚Üí prioritize marketing spend.

üî∏ Top Products


```python
product_revenue = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
product_revenue.head()

```


üìå Insight:
High-performing products should be stocked more.

üîπ Step 5: Time Series Analysis


```python
monthly_sales = df.groupby(['Year', 'Month'])['Revenue'].sum()
monthly_sales

monthly_sales.plot(title='Monthly Sales Trend')
plt.show()

```


üìå Business Insight:
Identifies seasonal peaks and slow periods.
