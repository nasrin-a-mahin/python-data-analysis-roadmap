Objective:

Learn to create new features from existing columns that capture business insights.



1️⃣ Import Libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```


2️⃣ Load Dataset


```python
df = pd.read_csv("retail_sales_cleaned.csv")
df.head()
```


3️⃣ Create Numeric Features
a) Total Revenue per Order


```python
df['Total_Revenue'] = df['Sales'] * df['Quantity']
```


b) Profit Margin


```python
df['Profit_Margin'] = df['Profit'] / df['Sales']
```


c) Discount Impact


```python
df['Revenue_after_Discount'] = df['Sales'] * (1 - df['Discount'])
```


4️⃣ Create Logical / Flag Features
High Value Order


```python
df['High_Value_Order'] = df['Sales'].apply(lambda x: 1 if x > 1000 else 0)
```


Profitability Flag


```python
df['Is_Profitable'] = df['Profit'].apply(lambda x: 1 if x>0 else 0)
```


5️⃣ Feature Interaction Example


```python
df['Revenue_per_Customer'] = df.groupby('Customer_ID')['Total_Revenue'].transform('sum')
```
