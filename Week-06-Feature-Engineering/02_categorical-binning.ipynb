Objective:

Transform categorical features into useful segments to reduce noise and improve analysis.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```


1️⃣ Load Data


```python
df = pd.read_csv("retail_sales_features.csv")
```


2️⃣ Combine Low-Frequency Categories


```python
top_categories = df['Product_Category'].value_counts().nlargest(3).index
df['Product_Category_Binned'] = df['Product_Category'].apply(
    lambda x: x if x in top_categories else 'Other'
)
```


3️⃣ Map Ordinal Features


```python
priority_map = {'High':3, 'Medium':2, 'Low':1}
df['Priority_Num'] = df['Priority'].map(priority_map)
```


4️⃣ Frequency Encoding (Optional)


```python
freq_encoding = df['Region'].value_counts().to_dict()
df['Region_Freq'] = df['Region'].map(freq_encoding)
```


5️⃣ Validate


```python
df['Product_Category_Binned'].value_counts()
df['Region_Freq'].head()

```


✅ Outcome:

Categorical features are now analysis-ready, with rare categories grouped and ordinal info encoded.
